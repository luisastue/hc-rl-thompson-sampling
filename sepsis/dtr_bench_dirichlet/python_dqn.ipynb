## Thompson sampling using gym DQN or PPO

```python
transition_models = [vi.transition_model(np.ones((vi.n_states, vi.n_actions, vi.n_states))) for _ in range(5)]
```

```python
nr_iter = 1_000
for i in range(5):
    tr = transition_models[i]
    mdps = [mdp.train_new_custom_mdp(tr, nr_iter, f"1000_{i}_{j}") for j in range(10)]
    rewards = [m.evaluate_policy(10000) for m in mdps]
    
    plt.figure()
    plt.hist(rewards, bins=10, alpha=0.7, edgecolor='black')
    plt.title(f"Histogram of Rewards for Transition Model {i}")
    plt.xlabel("Reward")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
    
```
