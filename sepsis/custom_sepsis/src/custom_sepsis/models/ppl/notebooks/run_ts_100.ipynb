```julia
include("../main.jl")
using PyCall;
using Revise;
using Gen;
using Serialization
using CairoMakie
sepsis_gym = pyimport("custom_sepsis");
```

```julia
models = Dict(
    :SimplePPL => [MCMCModel(:SimplePPL, choicemap(), [], []) for i in 1:3],
    :Softmax => [MCMCModel(:Softmax, choicemap(), [], []) for i in 1:3]
)
ts_runs = Dict(
    :SimplePPL => [PPTSRun("$(label_dict[model.type])-$i", model) for (i, model) in enumerate(models[:SimplePPL])],
    :Softmax => [PPTSRun("$(label_dict[model.type])-$i", model) for (i, model) in enumerate(models[:Softmax])],
)
```

```julia
for i in 1:50000
    for type in [:SimplePPL, :Softmax]
        for ts_run in ts_runs[type]
            if i < 100 && log2(i) % 1 == 0 || i % 100 == 0
                iterations = i == 1 ? 200 : 20
                run_ts_mcmc!(ts_run, i, iterations)
                policy = ts_run.policies[i]
                update_model!(ts_run.model, i, policy)
                save_ts_jld(ts_run)
                println("Finished episode ", ts_run.name, " ", i)
            end
        end
    end
end
```
